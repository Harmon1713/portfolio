```r
#Part 3 - create a stacked bar chart
```

```r
CarLocation = read_excel('C:/Users/jacki/OneDrive/Desktop/Entity/DSO104 Data Wrangling and Visualization/L4-p3-handson.xlsx')
```

```r
head(CarLocation)
```

```r
ggplot(data=CarLocation) +
  geom_bar(mapping = aes(x = Car, fill=Location)) + 
  ggtitle("Cars type by Location") +
  theme(plot.title=element_text(hjust=0.5)) +
  xlab("Car Type") +
  ylab("Frequency") 
```

```r
ggplot(data=CarLocation) +
  geom_bar(mapping = aes(x = Car, fill=Location), position = "fill") + 
  ggtitle("Car Type by Location (Same Height)") +
  theme(plot.title=element_text(hjust=0.5)) +
  xlab("Car Type") +
  ylab("Frequency")
```

```r
##Southwest is the location with the most cars bought. When put into a stacked graph with even tops, its easier to tell which car types were bought more/less than in each area.
```

```r
#Part 4 - Scatter Plot
```

```r
Crocodiles = read_excel('C:/Users/jacki/OneDrive/Desktop/Entity/DSO104 Data Wrangling and Visualization/L4-p4-handson.xlsx')
```

```r
head(Crocodiles)
```

```r
ggplot(data=Crocodiles, aes(x = BodyLength, y = HeadLength)) + 
geom_point() + 
geom_smooth(method='lm', se = TRUE) + 
ggtitle("Estuarine Crocodiles") + 
theme(plot.title=element_text(hjust=0.5)) + 
xlab("Length of Body") + 
ylab("Length of Head")
```

```r
##The longer the length of the body, the longer the length of the head.
```
